```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
```

```python
target_url4 = ("http://archive.ics.uci.edu/ml/machine-learning-databases/"
"glass/glass.data")
glass = pd.read_csv(target_url4,header=0,prefix='V')
```

```python
#glass.shape
glass.columns = ['Id','RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type']
glass.head()
```

```python
summary = glass.describe()
summary
```

```python
ncol = len(glass.columns)
glass_norm = copy.deepcopy(glass)
glass_norm = glass_norm.iloc[:,1:ncol] # 第一列是编号，不需要
ncol2 = len(glass_norm.columns)
summary2 = glass_norm.describe()
for i in range(ncol2):
    mean = summary2.iloc[1,i]
    sd = summary2.iloc[2,i]
    glass_norm.iloc[:,i] = (glass_norm.iloc[:,i] - mean) / sd
array = glass_norm.values # numpy的n维数组，ndarray
#type(array)
plt.boxplot(array)
plt.xlabel("Attribute Index")
plt.ylabel("Quartile Ranges - Glass Normalized")
plt.show()
```

```python
# 箱线图显示有相当的异常点，这是一个多分类的问题，在属性和类别之间不需要
# 存在连续性， 另外他的数据是非平衡性的，最多的类76个样本，最小的类9个样本
# 平行坐标图
ncols = glass_norm.shape[1]
nrows = glass_norm.shape[0]
for i in range(nrows):
    dataRow = glass_norm.iloc[i,1:ncols-1]
    labelColor = glass_norm.iloc[i,ncols-1] / 7.0
    dataRow.plot(color=plt.cm.RdYlBu(labelColor),alpha=0.5) # 平行坐标图
array = glass_norm.values # numpy的n维数组，ndarray
#type(array)
#plt.boxplot(array)
plt.xlabel("Attribute Index")
plt.ylabel("Quartile Ranges - Glass Normalized")
plt.show()
```
