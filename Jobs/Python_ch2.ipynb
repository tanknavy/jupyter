```python
# 属性和属性，或者属性和标签的相关程度可以由皮尔逊相关系数来量化
# 计算相关系数在是数值属性之间
dataRow2 = df2.iloc[:,1]
dataRow3 = df2.iloc[:,2]
dataRow21 = df2.iloc[:,20]
mean2 = 0.0;mean3 = 0.0;mean21 = 0.0
numElt = len(dataRow2)

for i in range(numElt):
    mean2 += dataRow2[i]/numElt # 求均值
    mean3 += dataRow3[i]/numElt
    mean21 += dataRow21[i]/numElt

var2 = 0.0;var3 = 0.0;var21 = 0.0
for i in range(numElt):
    var2 += (dataRow2[i] - mean2) ** 2 / numElt # 求方差
    var3 += (dataRow3[i] - mean3) ** 2 / numElt
    var21 += (dataRow21[i] - mean21) ** 2 / numElt

corr23 = 0.0; corr221 = 0.01
from math import sqrt
for i in range(numElt):
    corr23 += (dataRow2[i] - mean2) * (dataRow3[i] - mean3) / (sqrt(var2 * var3) * numElt)
    corr221 += (dataRow2[i] - mean2) * (dataRow21[i] - mean21) / (sqrt(var2 * var21) * numElt)

print(corr23,corr221)   
    
```

```python
# 热图（heat map）展示属性和标签的相关性
corMat = pd.DataFrame(df2.corr()) #皮尔森相关系数矩阵
plt.pcolor(corMat)
plt.show()
```

```python
# 回归问题，应该用不同的颜色来对应标签值的高低，实现标签的实数值到颜色值的映射
# 需要将标签的市属值研所到【0。0,1.0】区间
from math import exp
minRings = summary.iloc[3,7]
maxRings = summary.iloc[7,7]
nrows = len(df3.index)

for i in range(nrows):
    dataRow = df3.iloc[i,1:8]
    labelColor = (df3.iloc[i,8] - minRings) / (maxRings - minRings)
    dataRow.plot(color=plt.cm.RdYlBu(labelColor),alpha=0.5)
plt.xlabel("Attribute Index")
plt.ylabel("Quartile Ranges")
plt.show()
```

```python
# logit function
from math import exp
meanRings = summary.iloc[1,7]
sdRings = summary.iloc[2,7]
nrows = len(df3.index)
from math import exp

for i in range(nrows):
    dataRow = df3.iloc[i,1:8]
    normTarget = (df3.iloc[i,8] - meanRings) / sdRings
    labelColor = 1.0/(1.0 + exp(-normTarget)) # 对数转换，数值到(0,1)映射
    #labelColor = (df3.iloc[i,8] - minRings) / (maxRings - minRings)
    dataRow.plot(color=plt.cm.RdYlBu(labelColor),alpha=0.5)
plt.xlabel("Attribute Index")
plt.ylabel("Quartile Ranges")
plt.show()
```

```python
target_url3 = ("http://archive.ics.uci.edu/ml/machine-learning-databases/"
"wine-quality/winequality-red.csv")
wine = pd.read_csv(target_url3,header=0,sep=";")
wine.head()
```

```python
wine.shape
```

```python
import copy
wine_norm = copy.deepcopy(wine)
ncols = len(wine_norm.columns)
summary = wine.describe()
wine.head()
```

```python
for i in range(ncols):
    mean = summary.iloc[1,i]
    sd = summary.iloc[2,i]
    wine_norm.iloc[:,i] = (wine_norm.iloc[:,i] - mean) / sd
array = wine_norm.values
plt.boxplot(array)
plt.xlabel("Attribute Index")
plt.ylabel("Quartile Ranges - Wine Normalized")
plt.show()
```

```python
nrows = len(wine.index)
nDataCol = len(wine.columns) -1
tasteCol = len(summary.columns)
meanTaste = summary.iloc[1,tasteCol - 1]
sdTaste = summary.iloc[2,tasteCol - 1]
wine.head()
```

```python
# normalized label 
for i in range(nrows): # 
    dataRow = wine.iloc[i,1:nDataCol]
    normTarget = (wine.iloc[i,nDataCol] - meanTaste) / sdTaste
    labelColor = 1.0 / (1.0 + exp(-normTarget))
    dataRow.plot(color=plt.cm.RdYlBu(labelColor),alpha=0.5)
    #wine.iloc[i,:].plot(color=plt.cm.RdYlBu(labelColor),alpha=0.5)
```

```python
# normalized numeric values
# normalized label 
wineNorm = copy.deepcopy(wine) # 注意一定要deepcopy
for i in range(ncols):
    mean = summary.iloc[1,i]
    sd = summary.iloc[2,i]
    wineNorm.iloc[:,i] = (wineNorm.iloc[:,i] - mean) /sd # （0，1）归一化属性
for i in range(nrows): # 
    dataRow = wineNorm.iloc[i,1:nDataCol]
    #normTarget = (wineNorm.iloc[i,nDataCol] - meanTaste) / sdTaste
    normTarget = wineNorm.iloc[i,nDataCol] # 归一化
    labelColor = 1.0 / (1.0 + exp(-normTarget)) # log归一化标签值
    dataRow.plot(color=plt.cm.RdYlBu(labelColor),alpha=0.5)
    #wine.iloc[i,:].plot(color=plt.cm.RdYlBu(labelColor),alpha=0.5)
```
