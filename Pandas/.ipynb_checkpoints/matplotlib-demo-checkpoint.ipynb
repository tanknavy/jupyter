```python
import csv
from collections import Counter
```

```python
c = Counter(['Python','JavaScript','Python'])#计数器
c
c.update(['Scala','Python','Go']) #继续添加元素
c
```

```python
with open('e:/input/stackflow/lang_df.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file) #带有header的csv字典映射每行
    
    language_counter = Counter()
    
    #row = next(csv_reader) #迭代器
    for row in csv_reader: 
        language_counter.update(row['LanguageWorkedWith'].split(";"))
    #print(row)
    #print(row['LanguageWorkedWith'].split(";")) #['HTML/CSS', 'Java', 'JavaScript', 'Python']

languages = []
popularity = []
print(language_counter.most_common(15))
for item in language_counter.most_common(15): #item是tuple
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages.reverse() #列表反转
popularity.reverse()

plt.style.use("fivethirtyeight")

plt.barh(languages, popularity) #竖向和横向

plt.title('languages')#主题
plt.xlabel('language')
plt.ylabel('People use')

#plt.legend(['all devs','python']) #联想
plt.legend() #也可使用label中标识
#plt.tight_layout()

#plt.xticks(ticks=x_indexes, labels=ages_x)#x将索引恢复到原始标签

#plt.savefig('plot.png')#保存图片
plt.show()
```

```python
### pd处理csv
data = pd.read_csv('e:/input/stackflow/lang_df.csv')
data.dropna(inplace=True) #去除nan是float类型

ids = data['Respondent']
lang_responses = data['LanguageWorkedWith'] #Series类

language_counter = Counter()#计数器
    
    #row = next(csv_reader) #迭代器
for response in lang_responses: # Series
    #print(response)
    if not response == np.nan:
        language_counter.update(response.split(";")) #注意空字符

languages = []
popularity = []
print(language_counter.most_common(15))
for item in language_counter.most_common(15): #item是tuple
    languages.append(item[0])
    popularity.append(item[1])   
```

```python
languages.reverse() #列表反转
popularity.reverse()

plt.style.use("fivethirtyeight")

plt.barh(languages, popularity) #竖向和横向

plt.title('languages')#主题
plt.xlabel('language')
plt.ylabel('People use')

#plt.legend(['all devs','python']) #联想
plt.legend() #也可使用label中标识
#plt.tight_layout()

#plt.xticks(ticks=x_indexes, labels=ages_x)#x将索引恢复到原始标签

#plt.savefig('plot.png')#保存图片
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
from itertools import count
from matplotlib.animation import FuncAnimation
```

```python
from matplotlib.animation import FuncAnimation
plt.style.use('fivethirtyeight')

#实时数据
x_vals = []
y_vals = []

index = count()#itertools

def animate(i):
    x_vals.append(next(index))# 递增index
    y_vals.append(random.randint(0,5)) #随机数据
    
    plt.cla()#清除当前axis(重新设置)，否则会每次重新绘图
    plt.plot(x_vals, y_vals) #折线图
    
ani = FuncAnimation(plt.gcf(), animate, interval=1000)#每秒运行一次animate函数
    
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()#itertools

def animate(i):
    # use a real-time update csv file as data source
    data = pd.read_csv('e:/input/stackflow/time_series.csv')#此文件每秒更新一次
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']
    
    plt.cla()
    plt.plot(x, y1, label='Channel 1')
    plt.plot(x, y2, label='Channel 2')
    plt.legend(loc='upper left')
    plt.tight_layout()
    
    
ani = FuncAnimation(plt.gcf(), animate, interval=1000) #1秒更新一次绘图
    
plt.tight_layout()
plt.show()
```

```python
#读取csv文件
import csv,random,time

x_value = 0
total_1 = 1000
total_2 = 1000

fieldnames = ["x_value", "total_1", "total_2"]

# 先新建一个csv文件，写入header
with open('e:/input/stackflow/time_series.csv', 'w') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

#再打开这个csv文件每隔1秒追加一行数据
while True: 
    with open('e:/input/stackflow/time_series.csv', 'a') as csv_file:
        csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
         
        info = {
            "x_value" : x_value,
            "total_1" : total_1,
            "total_2" : total_2 }
        
        csv_writer.writerow(info)
        print(x_value, total_1, total_2)
        
        x_value +=1
        total_1 = total_1 + random.randint(-6,8)
        total_2 = total_2 + random.randint(-5,6)
    
    time.sleep(1)#暂停一秒
        
        
```
