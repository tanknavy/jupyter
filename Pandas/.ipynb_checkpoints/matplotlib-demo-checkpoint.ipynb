```python
import csv
from collections import Counter
```

```python
c = Counter(['Python','JavaScript','Python'])#
c
c.update(['Scala','Python','Go']) #
c
```

```python
with open('e:/input/stackflow/lang_df.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file) #headercsv
    
    language_counter = Counter()
    
    #row = next(csv_reader) #
    for row in csv_reader: 
        language_counter.update(row['LanguageWorkedWith'].split(";"))
    #print(row)
    #print(row['LanguageWorkedWith'].split(";")) #['HTML/CSS', 'Java', 'JavaScript', 'Python']

languages = []
popularity = []
print(language_counter.most_common(15))
for item in language_counter.most_common(15): #itemtuple
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages.reverse() #
popularity.reverse()

plt.style.use("fivethirtyeight")

plt.barh(languages, popularity) #

plt.title('languages')#
plt.xlabel('language')
plt.ylabel('People use')

#plt.legend(['all devs','python']) #
plt.legend() #label
#plt.tight_layout()

#plt.xticks(ticks=x_indexes, labels=ages_x)#x

#plt.savefig('plot.png')#
plt.show()
```

```python
### pdcsv
data = pd.read_csv('e:/input/stackflow/lang_df.csv')
data.dropna(inplace=True) #nanfloat

ids = data['Respondent']
lang_responses = data['LanguageWorkedWith'] #Series

language_counter = Counter()#
    
    #row = next(csv_reader) #
for response in lang_responses: # Series
    #print(response)
    if not response == np.nan:
        language_counter.update(response.split(";")) #

languages = []
popularity = []
print(language_counter.most_common(15))
for item in language_counter.most_common(15): #itemtuple
    languages.append(item[0])
    popularity.append(item[1])   
```

```python
languages.reverse() #
popularity.reverse()

plt.style.use("fivethirtyeight")

plt.barh(languages, popularity) #

plt.title('languages')#
plt.xlabel('language')
plt.ylabel('People use')

#plt.legend(['all devs','python']) #
plt.legend() #label
#plt.tight_layout()

#plt.xticks(ticks=x_indexes, labels=ages_x)#x

#plt.savefig('plot.png')#
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
plt.style.use('fivethirtyeight')

#
x_vals = []
y_vals = []

index = count()#itertools

def animate(i):
    x_vals.append(next(index))# index
    y_vals.append(random.randint(0,5)) #
    
    plt.cla()#axis()
    plt.plot(x_vals, y_vals) #
    
ani = FuncAnimation(plt.gcf(), animate, interval=1000)#animate
    
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()#itertools

def animate(i):
    # use a real-time update csv file as data source
    data = pd.read_csv('e:/input/stackflow/time_series.csv')#
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']
    
    plt.cla()
    plt.plot(x, y1, label='Channel 1')
    plt.plot(x, y2, label='Channel 2')
    plt.legend(loc='upper left')
    plt.tight_layout()
    
    
ani = FuncAnimation(plt.gcf(), animate, interval=1000) #1
    
plt.tight_layout()
plt.show()
```

```python
#csv
import csv,random,time

x_value = 0
total_1 = 1000
total_2 = 1000

fieldnames = ["x_value", "total_1", "total_2"]

# csvheader
with open('e:/input/stackflow/time_series.csv', 'w') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

#csv1
while True: 
    with open('e:/input/stackflow/time_series.csv', 'a') as csv_file:
        csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
         
        info = {
            "x_value" : x_value,
            "total_1" : total_1,
            "total_2" : total_2 }
        
        csv_writer.writerow(info)
        print(x_value, total_1, total_2)
        
        x_value +=1
        total_1 = total_1 + random.randint(-6,8)
        total_2 = total_2 + random.randint(-5,6)
    
    time.sleep(1)#
        
        
```
