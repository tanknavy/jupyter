# 数据清洗，却值处理

```python
people2 = {
    "first": ["alex","bob","carl","david", np.nan, None, 'NA'],
    "last":["Zhao","Qian","Cheng","Cheng", np.nan, np.nan, 'Missing'],
    "email":["alex@gmail.com","bob@126.com","carl@gmail.com", None, np.nan, 'anyone@gmail.com','NA'],
    "age":['33','55','63','36',None,None, 'Missing']
}
```

```python
df3 = pd.DataFrame(people2)
df3
```

```python
df3.dropna() #默认删除任何包含np.nan, None
```

```python
#df3.dropna() #默认删除任何包含np.nan, None的行
#df3.dropna(axis='index',how='any')#删除包含缺失值的行
#df3.dropna(axis='index',how='all')#删除全部都是缺失值的行
#df3.dropna(axis='columns',how='any')#删除全部都是缺失值的列
df3.dropna(axis='index',how='all',subset=['last','email'])#删除某些栏位(全部)包含缺失值的行
```

```python
df3.replace('NA',np.nan, inplace=True)#replace除了Series字典赋值，还可以对df元素替代
df3.replace('Missing',np.nan, inplace=True)#replace除了Series字典赋值，还可以对df元素替代
df3
```

```python
df3.isna()#每个元素是否na
```
