```python
# 预处理数据
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
# 特征工程，建立pipeline
def data_pipeline(data,output):
    """
    """
    data = cost_per_metric(data,output) if 'cost_per' in output else data[data[output] >0] #这里曝光预测值不能为负
    
    data = drop_columns(data,output,threshold=.5) # 删除太多空值的栏位
    data = data.dropna(axis='index') # 删除行如果包含空值
    data = create_other_buckets(data,threshold=.1) # categorial栏位中统一少数类别
    data = one_hot_encode(data) # 分类变量one-hot编码
    
    return data
    
    

def split_pipeline(data,output): #output代表y
    y,X = data.iloc[:,0],data.iloc[:,1:] # 第一列是响应变量y
    
    # 测试集/训练集
    X_train,X_test,y_train,y_test = train_test_split(data.drop([output],axis=1),data[output],test_size=.2,random_state=1) # 1代表列
    
    # 标准化后的测试集、训练集
    X_scaled,y_scaled,X_train_scaled,y_train_scaled,X_test_scaled,y_scaler = scale(X,y,X_train,y_train, X_test)
    
    return [X,y,X_train,y_train,X_test,y_test, X_scaled,y_scaled,X_train_scaled,y_train_scaled,X_test_scaled,y_scaler]


def cos_per_metric(data,output):
    metric = output.replace('cost_per', '') + 's' # 字符串取代
    data = data[data[metric] > 0]
    
    #DataFrame insert 栏位在指定的位置，值是两个栏位值的计算结果
    data.insert(0, output, [row['cost']/row[metric] for index,row in data.iterrows()]) # 逐行迭代，然后去栏位
    
    return data


# 如果栏位的空值率超过阈值，删除之
def drop_columns(data,output,threshold=0.5):
    rows = data[output].count()
    for column in list(data.columns):
        if data[column].count() < rows * threshold: # 有效值数量
            data = date.drop([column],axis=1)
        return data

    
# 稀少类别变量统一归入其它分桶
def create_other_buckets(data,threshold=0.1):
    categoricals = list(data.select_dtypes(include="object").columns) # 找到类别变量的栏位名
    
    for column in categoricals:
        results = data[column].count()
        groups = data.groupby([column])[column].count() # 按类别分组计数,pd.Series类型
        for bucket in groups.index: #每个类别
            if groups.loc[bucket] < results * threshold: #按照index名字定位
                data.loc[data[column] == bucket,column]# 凡是列值属于稀有类别的，稀有类别统一命名
    return data


# 类别独热编码
def one_hot_encode(data):
    categoricals = list(data.select_dtypes(include='object').columns) # 分类变量名
    for column in categoricals:
        if 'other' in data[column].unique(): # 如果列值包含other，即稀有分类
            data = pd.get_dummies(data,columns=[column],prefix=[column],drop_first=False) # k-1 of k个类比
            data = data.drop([column + '_other'],axis=1) # 删除稀有栏位
        else:
            data = pd.get_dummies(data,columns=[column],prefix=[column],drop_first=True) # True的含义第一类别变量设为NaN
    
    return data


# 数据标准化
def scale(X,y,X_train,y_train,X_test):
    X_scaler,y_scaler = StandardScaler(),StandardScaler() # 初始化scaler
    
    # X,y分别fit和transform原始数据
    X_scaled = X_scaler.fit_transform(X.values.astype(float)) #输入转为np.ndarray再转浮点型
    y_scaled = y_scaler.fit_transform(y.values.astype(float).reshape(-1,1)).flatten() # 为啥要转shape
    #y_scaled = y_scaler.fit_transform(y.values.astype(float)) # 为啥要转shape
    
    # 使用之前的scaler来transform测试数据，transform结果是np.ndarray,所以要加栏位名组成pd.DataFrame
    # 需要2d数据
    X_train_scaled = pd.DataFrame(data = X_scaler.transform(X_train.values.astype(float)) , columns=X.columns )
    #y_train_scaled = pd.DataFrame(dta = y_scaler.transform(y_train.values.astype(float)) )
    y_train_scaled = pd.DataFrame(data = y_scaler.transform(y_train.values.astype(float).reshape(-1,1)).flatten() )
    
    # 测试集标准化
    X_test_scaled = pd.DataFrame(data = X_scaler.transform(X_test.values.astype(float)), columns=X.columns)
    
    # 返回标准化后的数据以及y_scaler
    return [X_scaled,y_scaled,X_train_scaled,y_train_scaled,X_test_scaled,y_scaler] #还需要y_scaler用来转别
    
    
```
